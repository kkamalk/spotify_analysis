###Treating the outliers.

```python
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)
```

```python
iqr =  q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
```

```python
lower_bound
```

```python
upper_bound
```

```python
def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = (df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR))

   return outliers
```

```python
ol=find_outliers_IQR(data)
ol
```

```python
data.shape
```

```python
data_ol=data[~((data<(q1-1.5*iqr)) | (data>(q3+1.5*iqr))).any(axis=1)]
```
